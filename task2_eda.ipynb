In this task, I performed Exploratory Data Analysis on the cleaned Netflix dataset.
The goal was to understand patterns, trends, and insights using  Python visualizations such as Matplotlib and Seaborn.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

data = pd.read_csv("cleaned_netflix_data.csv")
display(data.info())
```

## Distribution of Numerical Columns

These histograms show how numerical values such as rating, popularity, vote count, vote average, budget, and revenue are spread across the dataset.
This helps identify skewness, spread, and presence of outliers.


```python
numeric_cols = ['rating', 'popularity', 'vote_count', 'vote_average', 'budget', 'revenue']

# Histograms
plt.figure(figsize=(14,8))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 3, i)
    sns.histplot(data[col], kde=True, bins=30)
    plt.title(f"Distribution of {col}")
plt.tight_layout()
plt.show()
```

## Top 10 Most Common Genres

This bar chart shows which genres appear most frequently on Netflix.
It gives a general idea about the type of content available.


```python
# Genres count
plt.figure(figsize=(10,5))
data['genres'].value_counts().head(10).plot(kind='bar', color='teal')
plt.title('Top 10 Most Common Genres')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```

## Movies vs TV Shows Count

This plot compares the number of Movies and TV Shows in the dataset.


```python
# Movie vs TV Show
plt.figure(figsize=(6,4))
sns.countplot(x='type', data=data, palette='pastel')
plt.title('Movies vs TV Shows Count')
plt.show()
```

## Number of Releases per Year

This line chart shows how the number of Netflix releases has changed over the years.
It helps us understand trends in content production.


```python
# Releases per year
plt.figure(figsize=(10,5))
data['release_year'].value_counts().sort_index().plot(kind='line', marker='o')
plt.title('Number of Releases per Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

## Correlation Heatmap

This heatmap shows the relationship between different numerical columns such as
budget, revenue, popularity, vote count, and rating.
Stronger correlations help us understand which variables influence each other.


```python
# Correlation heatmap

plt.figure(figsize=(8,6))
sns.heatmap(data.corr(numeric_only=True), cmap='coolwarm', annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

## Revenue Outlier Check

The boxplot is used to identify extreme values in the revenue column.
Outliers may represent exceptionally successful or unsuccessful movies.


```python
# Revenue outliers
plt.figure(figsize=(8,5))
sns.boxplot(x=data['revenue'])
plt.title('Revenue Outlier Check')
plt.show()
```

## Top 10 Countries Producing Netflix Content

This bar chart shows which countries contribute the most movies or TV shows
to Netflix.


```python
# Top countries
top_countries = data['country'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='viridis')
plt.title('Top 10 Producing Countries on Netflix')
plt.xlabel('Count')
plt.ylabel('Country')
plt.show()
```

# Conclusion
